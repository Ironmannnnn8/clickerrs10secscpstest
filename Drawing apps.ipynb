```python

import customtkinter as ctk
import time

window = ctk.CTk()
window.grid_propagate(False)


WIDTH=10000
HEIGHT=1000
timer=10

count=0


canvas=ctk.CTkCanvas(window,bg="black",width=WIDTH,height=HEIGHT)
canvas.grid()

def click():

  global count
  count+=1
  countbutton.configure(text=str(count))


  if timer==10:
      timers()#active kan timer sekali

  if timer==0:
      final_count=count/10
      resultbutton.configure(text=str(final_count) + "cps")







  window.update()

def timers():
    global timer
    if timer > 0:
        timer_text = '{:02d}'.format(timer)
        timebutton.configure(text=timer_text)
        timer -= 1
        window.after(1000, timers)  # Schedule next update in 1000 milliseconds (1 second)






button = ctk.CTkButton(window, text="Click here to play!", command=click, font=("Comic Sans", 10), fg_color="blue", bg_color="#0000FF",width=300,height=200)

button.place(anchor="center", relx=0.5, rely=0.5)


countbutton=ctk.CTkButton(window,text="",command=click,font=("Arial",10),fg_color='blue')
countbutton.place(anchor="center", relx=0.5, rely=0.1)

timebutton=ctk.CTkButton(window,text="",command=timers,font=("Arial",10),fg_color="blue")
timebutton.place(anchor="center", relx=0.5, rely=0.30)

resultbutton=ctk.CTkButton(window,text="",command=click,font=("Arial",50),fg_color="blue",width=100,height=100)
resultbutton.place(anchor="e", relx=0.8, rely=0.5, x=-30)


window.mainloop()


```
